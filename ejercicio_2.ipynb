```python
import yfinance as yf
import pandas as pd
import numpy as np

from finanzas_utils import calcular_retorno_diario, categorizar_volatilidad
```

```python
ticker1 = input("Ingrese el primer ticker: ").upper()
ticker2 = input("Ingrese el segundo ticker: ").upper()
```

```python
accion1 = yf.download(ticker1, period="1y")
accion2 = yf.download(ticker2, period="1y")
```

```python
# Retornos
retorno_anual_1 = calcular_retorno_diario(accion1['Close'][-1], accion1['Close'][0])
retorno_anual_2 = calcular_retorno_diario(accion2['Close'][-1], accion2['Close'][0])

retorno_mensual_1 = calcular_retorno_diario(accion1['Close'][-1], accion1['Close'][-21])
retorno_mensual_2 = calcular_retorno_diario(accion2['Close'][-1], accion2['Close'][-21])
```

```python
vol_anual_1 = accion1['Close'].pct_change().std() * np.sqrt(252) * 100
vol_anual_2 = accion2['Close'].pct_change().std() * np.sqrt(252) * 100

vol_mensual_1 = accion1['Close'].pct_change().std() * np.sqrt(21) * 100
vol_mensual_2 = accion2['Close'].pct_change().std() * np.sqrt(21) * 100
```

```python
print(f"{ticker1}: Volatilidad mensual {vol_mensual_1:.2f}% ({categorizar_volatilidad(vol_mensual_1)})")
print(f"{ticker2}: Volatilidad mensual {vol_mensual_2:.2f}% ({categorizar_volatilidad(vol_mensual_2)})")

print(f"{ticker1}: Volatilidad anual {vol_anual_1:.2f}% ({categorizar_volatilidad(vol_anual_1)})")
print(f"{ticker2}: Volatilidad anual {vol_anual_2:.2f}% ({categorizar_volatilidad(vol_anual_2)})")

print("\nMejor rendimiento anual:", ticker1 if retorno_anual_1 > retorno_anual_2 else ticker2)
print("Mejor rendimiento mensual:", ticker1 if retorno_mensual_1 > retorno_mensual_2 else ticker2)
```

```python
# ------------------------------------------------------------
# EJERCICIO 2 – COMPARADOR DE ACCIONES
#
# Descripción general:
# Este programa permite comparar dos acciones bursátiles
# utilizando datos financieros reales descargados desde Yahoo Finance.
#
# El programa:
# 1. Solicita al usuario dos tickers de acciones.
# 2. Descarga los precios históricos del último año.
# 3. Calcula el rendimiento mensual y anual.
# 4. Calcula la volatilidad mensual y anual.
# 5. Clasifica la volatilidad como Baja, Media o Alta.
# 6. Determina qué acción tuvo mejor rendimiento.
#
# Para el análisis se reutilizan funciones definidas previamente
# en el módulo finanzas_utils.py, promoviendo la modularización
# y reutilización del código.
# ------------------------------------------------------------


# ------------------------------------------------------------
# 1. IMPORTACIÓN DE LIBRERÍAS
# ------------------------------------------------------------
# yfinance: descarga datos financieros reales
# pandas: manejo de datos en formato tabla (DataFrame)
# numpy: cálculos matemáticos y estadísticos
# ------------------------------------------------------------

import yfinance as yf
import pandas as pd
import numpy as np


# ------------------------------------------------------------
# 2. IMPORTACIÓN DE FUNCIONES PROPIAS
# ------------------------------------------------------------
# Se importan las funciones creadas en el Ejercicio 1
# desde el archivo finanzas_utils.py
# ------------------------------------------------------------

from finanzas_utils import calcular_retorno_diario, categorizar_volatilidad


# ------------------------------------------------------------
# 3. ENTRADA DE DATOS DEL USUARIO
# ------------------------------------------------------------
# El usuario ingresa los tickers de las acciones a comparar.
# Se utiliza .upper() para asegurar que estén en mayúsculas,
# que es el formato estándar en los mercados financieros.
# ------------------------------------------------------------

ticker1 = input("Ingrese el primer ticker: ").upper()
ticker2 = input("Ingrese el segundo ticker: ").upper()


# ------------------------------------------------------------
# 4. DESCARGA DE DATOS FINANCIEROS
# ------------------------------------------------------------
# Se descargan los precios históricos diarios del último año
# para cada una de las acciones seleccionadas.
# ------------------------------------------------------------

accion1 = yf.download(ticker1, period="1y")
accion2 = yf.download(ticker2, period="1y")


# ------------------------------------------------------------
# VERIFICACIÓN 1:
# Los DataFrames deben contener columnas como:
# Open, High, Low, Close y Volume.
# ------------------------------------------------------------
# print(accion1.head())
# print(accion2.head())


# ------------------------------------------------------------
# 5. CÁLCULO DE RENDIMIENTOS
# ------------------------------------------------------------
# Se calcula el retorno porcentual:
# - Anual: primer precio vs último precio del año
# - Mensual: último precio vs precio de hace 21 días hábiles
# ------------------------------------------------------------

# Retorno anual
retorno_anual_1 = calcular_retorno_diario(
    accion1['Close'][-1], accion1['Close'][0]
)
retorno_anual_2 = calcular_retorno_diario(
    accion2['Close'][-1], accion2['Close'][0]
)

# Retorno mensual
retorno_mensual_1 = calcular_retorno_diario(
    accion1['Close'][-1], accion1['Close'][-21]
)
retorno_mensual_2 = calcular_retorno_diario(
    accion2['Close'][-1], accion2['Close'][-21]
)


# ------------------------------------------------------------
# VERIFICACIÓN 2:
# Los retornos deben ser valores razonables:
# positivos indican ganancia, negativos indican pérdida.
# ------------------------------------------------------------
# print(retorno_anual_1, retorno_anual_2)
# print(retorno_mensual_1, retorno_mensual_2)


# ------------------------------------------------------------
# 6. CÁLCULO DE VOLATILIDAD
# ------------------------------------------------------------
# La volatilidad se calcula como la desviación estándar de los
# retornos diarios y se escala a nivel anual y mensual.
#
# 252 días hábiles -> anual
# 21 días hábiles  -> mensual
# ------------------------------------------------------------

# Volatilidad anual
vol_anual_1 = accion1['Close'].pct_change().std() * np.sqrt(252) * 100
vol_anual_2 = accion2['Close'].pct_change().std() * np.sqrt(252) * 100

# Volatilidad mensual
vol_mensual_1 = accion1['Close'].pct_change().std() * np.sqrt(21) * 100
vol_mensual_2 = accion2['Close'].pct_change().std() * np.sqrt(21) * 100


# ------------------------------------------------------------
# VERIFICACIÓN 3:
# Las volatilidades deben ser positivas y coherentes con el
# mercado (normalmente entre 10% y 50%).
# ------------------------------------------------------------
# print(vol_anual_1, vol_anual_2)
# print(vol_mensual_1, vol_mensual_2)


# ------------------------------------------------------------
# 7. CLASIFICACIÓN E INTERPRETACIÓN DE LA VOLATILIDAD
# ------------------------------------------------------------
# Se utiliza la función categorizar_volatilidad para traducir
# valores numéricos en categorías cualitativas.
# ------------------------------------------------------------

print(f"{ticker1}: Volatilidad mensual {vol_mensual_1:.2f}% "
      f"({categorizar_volatilidad(vol_mensual_1)})")

print(f"{ticker2}: Volatilidad mensual {vol_mensual_2:.2f}% "
      f"({categorizar_volatilidad(vol_mensual_2)})")

print(f"{ticker1}: Volatilidad anual {vol_anual_1:.2f}% "
      f"({categorizar_volatilidad(vol_anual_1)})")

print(f"{ticker2}: Volatilidad anual {vol_anual_2:.2f}% "
      f"({categorizar_volatilidad(vol_anual_2)})")


# ------------------------------------------------------------
# 8. COMPARACIÓN DE RENDIMIENTOS
# ------------------------------------------------------------
# Se determina automáticamente qué acción tuvo mejor
# rendimiento mensual y anual.
# ------------------------------------------------------------

print("\nMejor rendimiento anual:",
      ticker1 if retorno_anual_1 > retorno_anual_2 else ticker2)

print("Mejor rendimiento mensual:",
      ticker1 if retorno_mensual_1 > retorno_mensual_2 else ticker2)


# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------
# El programa se considera correcto si:
# - No presenta errores de ejecución.
# - Descarga correctamente los datos.
# - Los valores obtenidos son coherentes.
# - Al cambiar los tickers, los resultados cambian.
# ------------------------------------------------------------
```
