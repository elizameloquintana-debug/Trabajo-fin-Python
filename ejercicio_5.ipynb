```python
# ------------------------------------------------------------
# EJERCICIO 5 – WEB SCRAPING AUTOMÁTICO (S&P 500)
#
# Descripción:
# En este ejercicio se realiza un proceso de Web Scraping
# sobre la página oficial de Wikipedia que contiene la
# información de las empresas que conforman el índice S&P 500.
#
# Los objetivos del ejercicio son:
# 1. Acceder a la página web del S&P 500.
# 2. Extraer automáticamente las tablas existentes.
# 3. Guardar la información en archivos CSV.
# 4. Identificar la empresa más antigua del índice.
# 5. Identificar la empresa que más veces ha salido y
#    la que más veces ha ingresado al índice.
#
# Este ejercicio demuestra el uso de Python para la
# automatización de la recolección de datos desde la web.
# ------------------------------------------------------------


# ------------------------------------------------------------
# 1. IMPORTACIÓN DE LIBRERÍAS
# ------------------------------------------------------------
# pandas: permite leer tablas HTML directamente desde una URL
# y manipular la información en DataFrames.
# ------------------------------------------------------------

import pandas as pd


# ------------------------------------------------------------
# 2. DEFINICIÓN DE LA URL
# ------------------------------------------------------------
# Se define la URL de Wikipedia que contiene las tablas
# con la información del índice S&P 500.
# ------------------------------------------------------------

url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"


# ------------------------------------------------------------
# 3. EXTRACCIÓN DE TABLAS DESDE LA WEB
# ------------------------------------------------------------
# La función read_html lee automáticamente todas las tablas
# encontradas en la página web y las almacena en una lista.
# ------------------------------------------------------------

tablas = pd.read_html(url)


# ------------------------------------------------------------
# VERIFICACIÓN 1:
# Se puede verificar cuántas tablas fueron encontradas.
# ------------------------------------------------------------
# print(len(tablas))


# ------------------------------------------------------------
# 4. SELECCIÓN DE LAS TABLAS RELEVANTES
# ------------------------------------------------------------
# La primera tabla contiene las empresas actuales del S&P 500.
# La segunda tabla contiene los cambios históricos del índice.
# ------------------------------------------------------------

sp500 = tablas[0]
cambios = tablas[1]


# ------------------------------------------------------------
# VERIFICACIÓN 2:
# Se inspeccionan las primeras filas de cada tabla.
# ------------------------------------------------------------
# print(sp500.head())
# print(cambios.head())


# ------------------------------------------------------------
# 5. GUARDADO DE LOS DATOS EN ARCHIVOS CSV
# ------------------------------------------------------------
# Se guardan las tablas extraídas en archivos CSV para su
# posterior análisis o reutilización.
# ------------------------------------------------------------

sp500.to_csv("sp500_companias.csv", index=False)
cambios.to_csv("sp500_cambios.csv", index=False)


# ------------------------------------------------------------
# VERIFICACIÓN 3:
# Los archivos CSV deben aparecer en la carpeta del proyecto.
# ------------------------------------------------------------


# ------------------------------------------------------------
# 6. IDENTIFICACIÓN DE LA EMPRESA MÁS ANTIGUA DEL ÍNDICE
# ------------------------------------------------------------
# Se busca la empresa con la fecha más antigua en la columna
# 'Date added', lo cual indica la primera incorporación al índice.
# ------------------------------------------------------------

empresa_antigua = sp500.loc[sp500['Date added'].idxmin()]


# ------------------------------------------------------------
# RESULTADO:
# Se muestra la información de la empresa más antigua.
# ------------------------------------------------------------

empresa_antigua


# ------------------------------------------------------------
# 7. EMPRESA CON MÁS SALIDAS Y MÁS ENTRADAS AL ÍNDICE
# ------------------------------------------------------------
# Se cuentan las veces que una empresa aparece en las columnas
# 'Removed' y 'Added' dentro de la tabla de cambios históricos.
# ------------------------------------------------------------

mas_salidas = cambios['Removed'].value_counts().idxmax()
mas_entradas = cambios['Added'].value_counts().idxmax()


# ------------------------------------------------------------
# 8. PRESENTACIÓN DE RESULTADOS
# ------------------------------------------------------------
# Se imprimen los resultados finales de forma clara.
# ------------------------------------------------------------

print("Empresa con más salidas del índice:", mas_salidas)
print("Empresa con más entradas al índice:", mas_entradas)


# ------------------------------------------------------------
# VALIDACIÓN FINAL DEL EJERCICIO
# ------------------------------------------------------------
# El ejercicio se considera correctamente implementado si:
# - Se accede correctamente a la página web.
# - Se extraen las tablas sin errores.
# - Se generan los archivos CSV.
# - Los resultados obtenidos son coherentes con la información
#   publicada en Wikipedia.
# ------------------------------------------------------------
```
