```python
# ------------------------------------------------------------
# EJERCICIO 3 – ANÁLISIS TÉCNICO: CRUCE DE MEDIAS (GOLDEN CROSS)
#
# Descripción:
# En este ejercicio se realiza un análisis técnico clásico
# conocido como "Golden Cross", el cual se basa en el cruce
# de dos medias móviles: una de corto plazo (50 días) y una
# de largo plazo (200 días).
#
# El Golden Cross ocurre cuando la media móvil de corto plazo
# cruza hacia arriba a la media móvil de largo plazo, lo que
# suele interpretarse como una posible señal alcista.
#
# Para este análisis se utilizan datos históricos reales de
# la acción Apple Inc. (AAPL) correspondientes a los últimos
# tres años.
# ------------------------------------------------------------


# ------------------------------------------------------------
# 1. IMPORTACIÓN DE LIBRERÍAS
# ------------------------------------------------------------
# yfinance: descarga de datos financieros reales
# matplotlib: visualización de series temporales
# ------------------------------------------------------------

import yfinance as yf
import matplotlib.pyplot as plt


# ------------------------------------------------------------
# 2. DESCARGA DE DATOS FINANCIEROS
# ------------------------------------------------------------
# Se descargan los precios históricos diarios de AAPL
# correspondientes a los últimos 3 años.
# ------------------------------------------------------------

data = yf.download("AAPL", period="3y")


# ------------------------------------------------------------
# VERIFICACIÓN 1:
# El DataFrame debe contener datos y la columna 'Close'
# ------------------------------------------------------------
# print(data.head())
# print(data.columns)


# ------------------------------------------------------------
# 3. CÁLCULO DE MEDIAS MÓVILES
# ------------------------------------------------------------
# Se calculan dos medias móviles simples (SMA):
# - MA50: media móvil de 50 días (corto plazo)
# - MA200: media móvil de 200 días (largo plazo)
#
# Estas medias se calculan sobre el precio de cierre.
# ------------------------------------------------------------

data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()


# ------------------------------------------------------------
# VERIFICACIÓN 2:
# Las primeras filas deben contener valores NaN debido
# al tamaño de la ventana de cálculo.
# ------------------------------------------------------------
# print(data[['MA50', 'MA200']].head(210))


# ------------------------------------------------------------
# 4. VISUALIZACIÓN DEL ANÁLISIS
# ------------------------------------------------------------
# Se genera un gráfico que muestra:
# - Precio de cierre
# - Media móvil de 50 días
# - Media móvil de 200 días
#
# El gráfico incluye título, leyenda y tamaño adecuado.
# ------------------------------------------------------------

plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Precio de Cierre')
plt.plot(data['MA50'], label='Media Móvil 50 días')
plt.plot(data['MA200'], label='Media Móvil 200 días')
plt.title('Análisis Golden Cross - AAPL')
plt.legend()
plt.show()


# ------------------------------------------------------------
# VERIFICACIÓN 3 (VISUAL):
# En el gráfico se deben observar posibles cruces entre
# las dos medias móviles.
# ------------------------------------------------------------


# ------------------------------------------------------------
# 5. DETECCIÓN DE PUNTOS DE CRUCE (GOLDEN CROSS)
# ------------------------------------------------------------
# Se identifica el punto donde:
# - MA50 pasa de estar por debajo de MA200
# - a estar por encima de MA200
#
# Esto se logra calculando la diferencia entre ambas medias
# y detectando el cambio de signo.
# ------------------------------------------------------------

data['Cruce'] = data['MA50'] - data['MA200']


# ------------------------------------------------------------
# VERIFICACIÓN 4:
# Cuando 'Cruce' pasa de negativo a positivo, ocurre
# un Golden Cross.
# ------------------------------------------------------------
# print(data[['MA50', 'MA200', 'Cruce']].tail())


# ------------------------------------------------------------
# 6. IDENTIFICACIÓN DE FECHAS DE GOLDEN CROSS
# ------------------------------------------------------------
# Se filtran los registros donde ocurre el cruce alcista.
# ------------------------------------------------------------

golden_cross = data[
    (data['Cruce'] > 0) & (data['Cruce'].shift(1) < 0)
]


# ------------------------------------------------------------
# 7. RESULTADOS
# ------------------------------------------------------------
# Se muestran las fechas y precios en los que ocurrió
# el Golden Cross.
# ------------------------------------------------------------

golden_cross[['Close']]


# ------------------------------------------------------------
# VALIDACIÓN FINAL DEL EJERCICIO
# ------------------------------------------------------------
# El ejercicio se considera correcto si:
# - Los datos se descargan correctamente.
# - Las medias móviles se calculan sin errores.
# - El gráfico muestra claramente las medias y el precio.
# - Se identifican correctamente los puntos de cruce.
#
# En caso de no observar cruces, se puede:
# - Cambiar el activo por uno más volátil (ej. TSLA)
# - Aumentar el período de análisis.
# ------------------------------------------------------------
```
